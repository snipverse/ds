```python
import json
```

```python
url= "https://jsonplaceholder.typicode.com/users"
```

```python
from urllib.request import urlopen
```

```python
page=urlopen(url)
```

```python
data=json.loads(page.read())
```

```python
Id=[]
Username=[]
Email=[]
```

```python
for item in data:
    if "id" in item.keys():
        Id.append(item['id'])
    else:
        Id.append("NA")
    if "username" in item.keys():
        Username.append(item['username'])
    else:
        Username.append("NA")
    if "email" in item.keys():
        Email.append(item['email'])
    else:
        Email.append("NA")
```

```python
import pandas as pd
df=pd.DataFrame()
```

```python
df["Id"]=Id
df["Username"]=Username
df["Email"]=Email
```
